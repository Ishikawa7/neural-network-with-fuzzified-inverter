```python
soggetto = "soggetto_2"
```

```python
# read csv
df = pd.read_csv("Datasets/" + soggetto + '.csv')
df
```

```python
# prepare empty dataframe
df_input_1 = pd.DataFrame()
df_input_2 = pd.DataFrame()
```

```python
SIGMA_MAX = 3
```

```python
# calculate bit one
df_input_1["SIGMA"] = df.SIGMA
df_input_1["bit_type"] = [1] * len(df)
df_input_1["bit"] = 1 / (1 + (df.SIGMA/SIGMA_MAX))
df_input_2["SIGMA"] = df.SIGMA
df_input_2["bit_type"] = [2] * len(df)
df_input_2["bit"] = 1 - df_input_1["bit"]
```

```python
df_input_1 = pd.concat([df_input_1, df[df.columns[1:]]], axis=1)
df_input_1
```

```python
df_input_2 = pd.concat([df_input_2, df[df.columns[1:]]], axis=1)
df_input_2
```

```python
df_input = pd.concat([df_input_1,df_input_2], axis=0)
df_input.sort_index(inplace=True)
df_input
```

```python
df_input.to_csv("Datasets/" + soggetto + '_input.csv', index=False)
```

```python
def calculate_target(dim, sigma, bit_type):
    if bit_type == 1:
        if dim > 0.5:
            return dim - sigma * ( (2 * dim -1)/ 2 )
        else:
            return dim + sigma * ( (1 - 2 * dim)/ 2 )
    else:
        return dim
```

```python
df_targets = pd.DataFrame()
for col in df_input.columns[3:]:
    list_col = []
    for i in range(0, len(df_input)):
        list_col.append(calculate_target(df_input.iloc[i][col],df_input.iloc[i].SIGMA, df_input.iloc[i].bit_type))
    df_targets[col] = list_col
df_targets
```

```python
df_targets.to_csv("Datasets/" + soggetto + '_targets.csv', index=False)
```
