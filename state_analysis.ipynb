```python
# read csv file soggetto_8_weigths_o.csv
df_o_8 = pd.read_csv('Datasets/soggetto_8_weights_o.csv')
# rename columns
df_o_8.columns = ["TrRi","RiTr","Plu","Ttra","Ttrb","Tr","MeTr","MoTr","RiRi","MoFo","Mmof","MoRi","MeRi","Ri","Comp"]
# rename index with the same name of the columns
df_o_8.index = df_o_8.columns
df_o_8
```

```python
# read csv file soggetto_1_weigths_o.csv
df_o_1 = pd.read_csv('Datasets/soggetto_1_weights_o.csv')
# rename columns
df_o_1.columns = ["TrRi","RiTr","Plu","Ttra","Ttrb","Tr","MeTr","MoTr","RiRi","MoFo","Mmof","MoRi","MeRi","Ri","Comp"]
# rename index with the same name of the columns
df_o_1.index = df_o_8.columns
df_o_1
```

```python
# read csv file soggetto_1_weigths_o.csv
df_o_7 = pd.read_csv('Datasets/soggetto_7_weights_o.csv')
# rename columns
df_o_7.columns = ["TrRi","RiTr","Plu","Ttra","Ttrb","Tr","MeTr","MoTr","RiRi","MoFo","Mmof","MoRi","MeRi","Ri","Comp"]
# rename index with the same name of the columns
df_o_7.index = df_o_8.columns
df_o_7
```

```python
# plot heatmap with plotly express
fig = px.imshow(df_o_8, color_continuous_midpoint=0, color_continuous_scale = 'RdBu_r', zmin=-5, zmax=5, title="Soggetto_8", text_auto=".2", aspect="Auto")
fig.show()

fig = px.imshow(df_o_1, color_continuous_midpoint=0, color_continuous_scale = 'RdBu_r', zmin=-5, zmax=5, title= "Soggetto_1", text_auto=".2", aspect="Auto")
fig.show()

fig = px.imshow(df_o_7, color_continuous_midpoint=0, color_continuous_scale = 'RdBu_r', zmin=-5, zmax=5, title= "Soggetto_7", text_auto=".2", aspect="Auto")
fig.show()
```

```python
fig = px.bar([df_o_8.std(axis=1), df_o_1.std(axis=1), df_o_7.std(axis=1)], barmode='group')
fig.update_traces(marker_color=['red', 'blue',"green"])
# remove legend
fig.update_layout(showlegend=False)
# xticks labels = soggetto_8, soggetto_1, soggetto_7
fig.update_xaxes(ticktext=['soggetto_8', 'soggetto_1', 'soggetto_7'])
# yticks labels = std
fig.update_yaxes(title_text='std')
fig.show()
```

```python
np_8 = df_o_8.to_numpy()
np_1 = df_o_1.to_numpy()
np_7 = df_o_7.to_numpy()

# find the eigenvalues and eigenvectors of the matrix
eigvals_8, eigvecs_8 = np.linalg.eig(np_8)
eigvals_1, eigvecs_1 = np.linalg.eig(np_1)
eigvals_7, eigvecs_7 = np.linalg.eig(np_7)
```

```python
df_evolve_8 = pd.DataFrame()
df_evolve_1 = pd.DataFrame()
df_evolve_7 = pd.DataFrame()
for i in range(0,20):
    np_8 = np.matmul(np_8, np_8)
    sum_np_8 = np.sum(np_8, axis=1)
    df_evolve_8 = df_evolve_8.append(pd.Series(sum_np_8), ignore_index=True)
    np_8 = np.divide(np_8, sum_np_8)

    np_1 = np.matmul(np_1, np_1)
    sum_np_1 = np.sum(np_1, axis=1)
    df_evolve_1 = df_evolve_1.append(pd.Series(sum_np_1), ignore_index=True)
    np_1 = np.divide(np_1, sum_np_1)

    np_7 = np.matmul(np_7, np_7)
    sum_np_7 = np.sum(np_7, axis=1)
    df_evolve_7 = df_evolve_7.append(pd.Series(sum_np_7), ignore_index=True)
    np_7 = np.divide(np_7, sum_np_7)

df_evolve_8.columns = ["TrRi","RiTr","Plu","Ttra","Ttrb","Tr","MeTr","MoTr","RiRi","MoFo","Mmof","MoRi","MeRi","Ri","Comp"]
df_evolve_7.columns = ["TrRi","RiTr","Plu","Ttra","Ttrb","Tr","MeTr","MoTr","RiRi","MoFo","Mmof","MoRi","MeRi","Ri","Comp"]
df_evolve_1.columns = ["TrRi","RiTr","Plu","Ttra","Ttrb","Tr","MeTr","MoTr","RiRi","MoFo","Mmof","MoRi","MeRi","Ri","Comp"]
```

```python
fig = px.line(df_evolve_8, title='Soggetto 8')
fig.show()
fig = px.line(df_evolve_1, title='Soggetto 1')
fig.show()
fig = px.line(df_evolve_7, title='Soggetto 7')
fig.show()
```

```python
# fourier transform of df_evolve_8 and df_evolve_1
fft_8 = np.fft.fft(df_evolve_8)
fft_1 = np.fft.fft(df_evolve_1)
fft_7 = np.fft.fft(df_evolve_7)
```

```python
df_aplitudine_8 = pd.DataFrame(np.abs(fft_8), columns=df_evolve_8.columns)
df_aplitudine_1 = pd.DataFrame(np.abs(fft_1), columns=df_evolve_1.columns)
df_aplitudine_7 = pd.DataFrame(np.abs(fft_7), columns=df_evolve_7.columns)
```

```python
df_freq_8 = pd.DataFrame(np.angle(fft_8), columns=df_evolve_8.columns)
df_freq_1 = pd.DataFrame(np.angle(fft_1), columns=df_evolve_1.columns)
df_freq_7 = pd.DataFrame(np.angle(fft_7), columns=df_evolve_7.columns)
```

```python
# plot ampliture of fourier transform
fig = px.bar(df_aplitudine_8, title='Soggetto 8 amplitude', barmode='group')
fig.show()
# plot ampliture of fourier transform
fig = px.bar(df_aplitudine_1, title='Soggetto 1 amplitude', barmode='group')
fig.show()
# plot ampliture of fourier transform
fig = px.bar(df_aplitudine_7, title='Soggetto 7 amplitude', barmode='group')
fig.show()
```

```python
# plot frequency of fourier transform
fig = px.bar(df_freq_8, title='Soggetto 8 frequency', barmode='group')
fig.show()
# plot frequency of fourier transform
fig = px.bar(df_freq_1, title='Soggetto 2 frequency', barmode='group')
fig.show()
# plot frequency of fourier transform
fig = px.bar(df_freq_7, title='Soggetto 7 frequency', barmode='group')
fig.show()
```

```python
# calculate delta frequency and plot it
fig = px.bar(df_freq_8.diff(), title='Soggetto 8 delta frequency', barmode='group')
fig.show()
# calculate delta frequency and plot it
fig = px.bar(df_freq_1.diff(), title='Soggetto 1 delta frequency', barmode='group')
fig.show()
# calculate delta frequency and plot it
fig = px.bar(df_freq_7.diff(), title='Soggetto 7 delta frequency', barmode='group')
fig.show()
```
