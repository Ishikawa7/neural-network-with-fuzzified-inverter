```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
import plotly
import plotly.express as px
import plotly.graph_objects as go
sns.set()
```

```python
# read csv file soggetto_8_weigths_o.csv
df_o_8 = pd.read_csv('soggetto_8_weights_o.csv')
df_o_8
```

```python
# read csv file soggetto_8_weigths_o.csv
df_o_1 = pd.read_csv('soggetto_1_weights_o.csv')
df_o_1
```

```python
# plot heatmap with plotly express
fig = px.imshow(df_o_8, color_continuous_midpoint=0, color_continuous_scale = 'RdBu_r', zmin=-5, zmax=5, title="Soggetto_8")
fig.show()

fig = px.imshow(df_o_1, color_continuous_midpoint=0, color_continuous_scale = 'RdBu_r', zmin=-5, zmax=5, title= "Soggetto_1")
fig.show()
```

```python
fig = px.bar([df_o_8.std(axis=1), df_o_1.std(axis=1)], barmode='group')
fig.update_traces(marker_color=['red', 'blue'])
fig.show()
```

```python
np_8 = df_o_8.to_numpy()
np_1 = df_o_1.to_numpy()

# find the eigenvalues and eigenvectors of the matrix
eigvals_8, eigvecs_8 = np.linalg.eig(np_8)
eigvals_1, eigvecs_1 = np.linalg.eig(np_1)
```

```python
eigvals_8
```

```python
eigvals_1
```

```python
df_evolve_8 = pd.DataFrame()
df_evolve_1 = pd.DataFrame()
for i in range(0,51):
    np_8 = np.matmul(np_8, np_8)
    sum_np_8 = np.sum(np_8, axis=1)
    df_evolve_8 = df_evolve_8.append(pd.Series(sum_np_8), ignore_index=True)
    np_8 = np.divide(np_8, sum_np_8)

    np_1 = np.matmul(np_1, np_1)
    sum_np_1 = np.sum(np_1, axis=1)
    df_evolve_1 = df_evolve_1.append(pd.Series(sum_np_1), ignore_index=True)
    np_1 = np.divide(np_1, sum_np_1)
```

```python
fig = px.line(df_evolve_8, title='Soggetto 8')
fig.show()
fig = px.line(df_evolve_1, title='Soggetto 1')
fig.show()
```

```python
# fourier transform of df_evolve_8 and df_evolve_1
df_evolve_8_fft = np.fft.fft(df_evolve_8)
df_evolve_1_fft = np.fft.fft(df_evolve_1)
```

```python
# plot ampliture of fourier transform
fig = px.bar(np.abs(df_evolve_8_fft), title='Soggetto 8 amplitude', barmode='group')
fig.show()
# plot frequency of fourier transform
fig = px.bar(np.angle(df_evolve_8_fft), title='Soggetto 8 frequency', barmode='group')
fig.show()
```

```python
# plot ampliture of fourier transform
fig = px.bar(np.abs(df_evolve_1_fft), title='Soggetto 1 amplitude', barmode='group')
fig.show()
# plot frequency of fourier transform
fig = px.bar(np.angle(df_evolve_1_fft), title='Soggetto 2 frequency', barmode='group')
fig.show()
```
